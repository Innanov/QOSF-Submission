### Solution:

```python
from qiskit import*
from qiskit.visualization import plot_histogram
```

```python
#Circuit:
qc = QuantumCircuit(4)
for i in range(4):
    qc.h(i)
qc.x(1)
qc.h(1)   
qc.ccx(0,3,1)
qc.ccx(2,3,1)
qc.h(1)
qc.x(1)
for qubit in range(4):
    qc.h(qubit)
for qubit in range(4):
    qc.x(qubit)
qc.h(3)
qc.mct(list(range(3)),3) 
qc.h(3)
for qubit in range(4):
    qc.x(qubit)
for qubit in range(4):
    qc.h(qubit)
#_________________    
qc.x(1)
qc.h(1)   
qc.ccx(0,3,1)
qc.ccx(2,3,1)
qc.h(1)
qc.x(1)
for qubit in range(4):
    qc.h(qubit)
for qubit in range(4):
    qc.x(qubit)
qc.h(3)
qc.mct(list(range(3)),3)  
qc.h(3)
for qubit in range(4):
    qc.x(qubit)
for qubit in range(4):
    qc.h(qubit)    

qc.measure_all()
qc.draw(output='mpl')
```

```python
#Results
simulator = Aer.get_backend('aer_simulator')
counts = execute(qc, backend=simulator).result().get_counts(qc)

plot_histogram(counts)
```
